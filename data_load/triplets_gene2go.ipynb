```python
import pickle
import os


data_path = ''
file_name = 'gene2go_all.pkl'
entity2id_outputfile = 'entity2id.txt'
output_file = 'gene_go_triplets.txt'


try:
    with open(os.path.join(data_path, file_name), 'rb') as f:
        gene2go = pickle.load(f)
except FileNotFoundError:
    print(f"File {file_name} not found in {data_path}")
    gene2go = None  
except pickle.UnpicklingError:
    print("Error loading pickle file. It may be corrupted.")
    gene2go = None
    

if gene2go:
    
    entities = set()
    relations = set()
    for gene, go_terms in gene2go.items():
        entities.add(gene)
        for go_term in go_terms:
            entities.add(go_term)
            relations.add('annotates')
    try:
        
        with open(os.path.join(data_path, entity2id_outputfile), 'w') as f:
            
            for i, entity in enumerate(entities):
                f.write(f"{entity}\t{i}\n")
        print(f"Entities successfully saved to {entity2id_outputfile}")
    except IOError:
        print(f"Error writing to file {entity2id_outputfile}")
        
```

```python
import pickle
import os


data_path = ''
file_name = 'gene2go_all.pkl'
entity2id_file = 'entity2id.txt'
output_file = 'gene_go_triplets.txt'


try:
    with open(os.path.join(data_path, file_name), 'rb') as f:
        gene2go = pickle.load(f)
except FileNotFoundError:
    print(f"File {file_name} not found in {data_path}")
    gene2go = None
except pickle.UnpicklingError:
    print("Error loading pickle file. It may be corrupted.")
    gene2go = None
    

if gene2go:
    
    try:
        with open(os.path.join(data_path, entity2id_file), 'r') as f:
            entity2id = {line.split('\t')[0]: line.split('\t')[1].strip() for line in f}
    except FileNotFoundError:
        print(f"File {entity2id_file} not found in {data_path}")
        entity2id = None
        
    
    if entity2id:
        try:
            
            with open(os.path.join(data_path, output_file), 'w') as out_file:
                for gene, go_terms in gene2go.items():
                    
                    for go_term in go_terms:
                        out_file.write(f"{entity2id[gene]}\tannotates\t{entity2id[go_term]}\n")
            print(f"Triples successfully saved to {output_file}")
        except IOError:
            print(f"Error writing to file {output_file}")
    else:
        print("No entity to id data to process.")
    
```

```python
import os


data_path = ''
file = 'go_go_triplets.txt'
entity2id_file = 'entity2id.txt'


try:
    with open(os.path.join(data_path, file), 'r') as f:
        data = f.readlines()
except FileNotFoundError:
    print(f"File {file} not found in {data_path}")
    data = None
    

if data:
    
    try:
        with open(os.path.join(data_path, entity2id_file), 'r') as f:
            entity2id = {line.split('\t')[0]: line.split('\t')[1].strip() for line in f}
    except FileNotFoundError:
        print(f"File {entity2id_file} not found in {data_path}")
        entity2id = None
        
    
    if entity2id:
        
        entities = set()
        for line in data:
            head, relation, tail = line.strip().split('\t')
            entities.add(head)
            entities.add(tail)
        
        new_entities = [entity for entity in entities if entity not in entity2id]
        
        try:
            
            with open(os.path.join(data_path, entity2id_file), 'a') as f:
                
                for i, entity in enumerate(new_entities):
                    f.write(f"{entity}\t{i+len(entity2id)}\n")
            print(f"New entities successfully saved to {entity2id_file}")
        except IOError:
            print(f"Error writing to file {entity2id_file}")
    else:
        print("No entity to id data to process.")
```

# Splitting the data into train, validation, and test sets


```python


import os
import random


data_path = ''
file = 'triplets.txt'
train_file = 'train.txt'
valid_file = 'valid.txt'
test_file = 'test.txt'


try:
    with open(os.path.join(data_path, file), 'r') as f:
        data = f.readlines()
except FileNotFoundError:
    print(f"File {file} not found in {data_path}")
    data = None
    

if data:
    try:
        
        with open(os.path.join(data_path, train_file), 'w') as train_f, \
             open(os.path.join(data_path, valid_file), 'w') as valid_f, \
             open(os.path.join(data_path, test_file), 'w') as test_f:
            
            random.shuffle(data)
            
            train_end = int(0.8 * len(data))
            valid_end = int(0.9 * len(data))
            
            train_f.writelines(data[:train_end])
            valid_f.writelines(data[train_end:valid_end])
            test_f.writelines(data[valid_end:])
        print(f"Data successfully split and saved to {train_file}, {valid_file} and {test_file}")
    except IOError:
        print(f"Error writing to files {train_file}, {valid_file} or {test_file}")

    
```

```python
import os


data_path = ''
file = 'GO_IDs_and_Namespaces.csv'
entity2id_file = 'entity2id.txt'
output_file = 'GO_namespace.txt'


try:
    with open(os.path.join(data_path, file), 'r') as f:
        data = f.readlines()
except FileNotFoundError:
    print(f"File {file} not found in {data_path}")
    data = None
    

if data:
    
    try:
        with open(os.path.join(data_path, entity2id_file), 'r') as f:
            entity2id = {line.split('\t')[0]: line.split('\t')[1].strip() for line in f}
    except FileNotFoundError:
        print(f"File {entity2id_file} not found in {data_path}")
        entity2id = None
        
    
    if entity2id:
        try:
            
            with open(os.path.join(data_path, output_file), 'w') as out_file:
                for line in data:
                    go_id, namespace = line.strip().split(',')
                    if go_id in entity2id:
                        out_file.write(f"{entity2id[go_id]}\t{namespace}\n")
            print(f"GO term namespaces successfully saved to {output_file}")
        except IOError:
            print(f"Error writing to file {output_file}")
    else:
        print("No entity to id data to process.")
```

```python

import os


data_path = ''
entity_attributes_file = 'entity_attributes.txt'
go_namespace_file = 'GO_namespace.txt'
entity2id_file = 'entity2id.txt'
output_file = 'all_entity.csv'



try:
    with open(os.path.join(data_path, entity_attributes_file), 'r') as f:
        entity_attributes = {line.split('\t')[0]: line.split('\t')[1].strip() for line in f}
except FileNotFoundError:
    print(f"File {entity_attributes_file} not found in {data_path}")
    entity_attributes = None
    

try:
    with open(os.path.join(data_path, go_namespace_file), 'r') as f:
        go_namespace = {line.split('\t')[0]: line.split('\t')[1].strip() for line in f}
except FileNotFoundError:
    print(f"File {go_namespace_file} not found in {data_path}")
    go_namespace = None
    

try:
    with open(os.path.join(data_path, entity2id_file), 'r') as f:
        entity2id = {line.split('\t')[0]: line.split('\t')[1].strip() for line in f}
except FileNotFoundError:
    print(f"File {entity2id_file} not found in {data_path}")
    entity2id = None
    


if entity_attributes and entity2id:
    try:
        
        with open(os.path.join(data_path, output_file), 'w') as f:
            f.write("entity, id, type,namespace\n")
            for entity, id in entity2id.items():
                f.write(f"{entity},{id}, {entity_attributes[id]},{go_namespace[id] if id in go_namespace else 'None'}\n")
        print(f"Entity attributes successfully saved to {output_file}")
    except IOError:
        print(f"Error writing to file {output_file}")


```
